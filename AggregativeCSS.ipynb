```python
def calculate_difference_list(lst):
    diff_lst = []
    for i in range(len(lst)-1):
        diff_lst.append(lst[i+1] - lst[i])
    return diff_lst

alabama=merged_df[merged_df["state"]=="Alabama"]
listx=calculate_difference_list(alabama["overall_amount_spent"].tolist())
listy=calculate_difference_list(alabama["deaths"].tolist())
```

```python
import statsmodels.api as sm

ols_fit_overall=pd.DataFrame()
states=[]
rsq=[]
adjrsq=[]
pval=[]
tval=[]
for i in set(merged_df["state"]):
  states_df=merged_df[merged_df["state"]==i]
  listx=calculate_difference_list(states_df["overall_amount_spent"].tolist())
  listy=calculate_difference_list(states_df["deaths"].tolist())
  listx = sm.add_constant(listx)
  ols = sm.OLS(listy,listx,missing = 'drop')
  ols_result = ols.fit()
  ols_new_results=ols_result.get_robustcov_results()
  print(ols_new_results.summary())
  states.append(i)
  rsq.append(ols_new_results.rsquared)
  adjrsq.append(ols_new_results.rsquared_adj)
  pval.append(ols_new_results.pvalues[1])
  tval.append(ols_new_results.tvalues[1])

ols_fit_overall["State"]=states
ols_fit_overall["R^2"]=rsq
ols_fit_overall["Adj_R^2"]=adjrsq
ols_fit_overall["pval"]=pval
ols_fit_overall["tval"]=tval
```

Granger Causality

```python
from statsmodels.tsa.stattools import grangercausalitytests

def granger_causality(df, col1, col2,state,maxlag):
    df=df[df["state"]==state]
    data = df[[col1, col2]].values
    results = grangercausalitytests(data, maxlag=maxlag, verbose=False)
    return results
```
